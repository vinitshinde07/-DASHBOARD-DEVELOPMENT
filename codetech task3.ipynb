```python
from google.colab import files
uploaded = files.upload()
import dash
from dash import dcc, html, Input, Output
import pandas as pd
import plotly.express as px

# Load the dataset
file_path = 'diabetes.csv'
df = pd.read_csv(file_path)

# Inspect column names
columns = df.columns

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the dashboard
app.layout = html.Div([
    html.H1("Diabetes Dataset Dashboard", style={'textAlign': 'center'}),
    html.Div([
        html.Label("Select a Feature to Visualize:"),
        dcc.Dropdown(
            id='feature-selector',
            options=[{'label': col, 'value': col} for col in columns if df[col].dtype in ['float64', 'int64']],
            value=columns[0],
        ),
        html.Label("Select a Target Category (Outcome):"),
        dcc.RadioItems(
            id='outcome-filter',
            options=[
                {'label': 'All', 'value': 'All'},
                {'label': 'Diabetic (1)', 'value': 1},
                {'label': 'Non-Diabetic (0)', 'value': 0},
            ],
            value='All',
            inline=True,
        ),
    ], style={'padding': '20px'}),
    dcc.Graph(id='feature-distribution-chart'),
])

# Callback to update the graph based on user selections
@app.callback(
    Output('feature-distribution-chart', 'figure'),
    [Input('feature-selector', 'value'),
     Input('outcome-filter', 'value')]
)
def update_chart(selected_feature, selected_outcome):
    filtered_df = df.copy()
    if selected_outcome != 'All':
        filtered_df = filtered_df[filtered_df['Outcome'] == selected_outcome]

    fig = px.histogram(
        filtered_df,
        x=selected_feature,
        color='Outcome',
        barmode='overlay',
        title=f"Distribution of {selected_feature} by Outcome",
        labels={'Outcome': 'Diabetes Outcome'}
    )
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
